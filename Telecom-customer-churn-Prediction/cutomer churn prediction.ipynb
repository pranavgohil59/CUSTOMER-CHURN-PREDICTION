```python
# DATA CLEANING

# DROP : DeviceProtection	TechSupport	StreamingTV	StreamingMovies  OnlineSecurity	OnlineBackup	DeviceProtection	TechSupport	
# 
```

```python
df1=df.drop(['DeviceProtection','TechSupport','StreamingTV','StreamingMovies','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','MultipleLines','PaperlessBilling'],axis=1)
```

```python
df1.drop('gender',axis=1,inplace=True)
```

```python
# PREPROCSSING

#  ONE HOT ENCODING : GENDER, DEPENDENTS , PHONESERVICE, INTERNETSERVICE, PATMENTMETHOD
#  ORDINAL ENCODING : PARNET , CONTRACT
#  LABEL ENCODING: CHURN
```

```python
df['Churn'] =df['Churn'].apply(lambda x : 1 if x=='Yes' else 0 )
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,accuracy_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder , OrdinalEncoder, LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
x = df1.drop('Churn',axis=1)
y = df1['Churn']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2 ,random_state=42)
```

```python
dt = DecisionTreeClassifier(
    criterion='entropy',
    splitter='random',
    max_depth=10,
    min_samples_split = 5,
    max_features ='sqrt'
)
rfc = RandomForestClassifier(
    n_estimators=200,
    max_depth=15,
    min_samples_split=5,
    min_samples_leaf=2,
    max_features= 'sqrt'
)
```

```python
transformer = ColumnTransformer(transformers=[
    ('ohe', OneHotEncoder(drop='first'),['Dependents' , 'PhoneService', 'InternetService', 'PaymentMethod']),
    ('ordinal', OrdinalEncoder(categories=[['Yes','No'],['Month-to-month','One year','Two year']]),['Partner','Contract']), 
],remainder='passthrough')
```

```python
le = LabelEncoder()
```

```python
y_train=le.fit_transform(y_train)
```

```python
y_test = le.transform(y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'criterion':['gini','Entropy'],
    'splitter':['Best','random'],
    'max_depth':[2,4,6,8,10,12] ,
    'min_samples_split' : [1,2,3,4,5,6,7],
    'max_features' : [2,3,4,'sqrt']
}
grid = GridSearchCV(dt,params,cv=5,scoring='accuracy')
```

```python
params = {
        'n_estimators': [50, 100, 150],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt']
}
grid1 = GridSearchCV(rfc,params,cv=5,scoring='accuracy')
```

```python
pipe = Pipeline([
    ('transformer',transformer),
    ('dtree',grid1)
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test,y_pred))
print("ROC-AUC:", roc_auc_score(y_test, pipe.predict_proba(x_test)[:, 1]))
```
